````python
# TODO: متد زیر را برای نشان دادن KDE Plot تکمیل و متغیر های ذکر شده را بررسی کنید.
def plot_numeric_distributions(df, numeric_columns, target_col='Churn Label'):
    """
    رسم نمودارهای KDE برای متغیرهای عددی با تفکیک بر اساس متغیر هدف.
    این تابع توزیع متغیرهای عددی را برای هر مقدار از متغیر هدف (مانند 'Churn Label')
    در یک شبکه ۲×۲ نمایش می‌دهد.
    پارامترها:
    df (DataFrame): دیتافریم شامل داده‌ها
    numeric_cols (list): لیست ستون‌های عددی برای رسم نمودار
    target_col (str): نام ستون هدف برای دسته‌بندی (پیش‌فرض 'Churn Label')
    """
    pass
````

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def plot_numeric_distributions(df, numeric_columns, target_col='Churn Label'):
   
    if target_col not in df.columns:
        raise ValueError(f"ستون '{target_col}' در دیتافریم وجود ندارد.")
    
    n_cols = 2
    n_rows = (len(numeric_columns) // n_cols + (1 if len(numeric_columns) % n_cols != 0 else 0))
    
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 5*n_rows))
    
    if n_rows == 1:
        axes = np.array([axes])
    
    axes = axes.flatten()
    
    for i, col in enumerate(numeric_columns):
        if col in df.columns:
            sns.kdeplot(data=df, x=col, hue=target_col, ax=axes[i], fill=True, common_norm=False,
                       palette={'Yes': 'red', 'No': 'green'}, alpha=0.6)
            
            axes[i].set_title(f'توزیع {col} بر اساس {target_col}', fontsize=12)
            axes[i].set_xlabel(col, fontsize=10)
            axes[i].set_ylabel('چگالی', fontsize=10)
            axes[i].grid(True, linestyle='--', alpha=0.6)
            
            for value in df[target_col].unique():
                mean_val = df[df[target_col]==value][col].mean()
                axes[i].axvline(mean_val, color='red' if value=='Yes' else 'green', 
                                linestyle='--', linewidth=1.5,
                                label=f'میانگین {value}: {mean_val:.2f}')
            
            axes[i].legend()
        else:
            print(f"ستون '{col}' در دیتافریم وجود ندارد.")
    
    for j in range(i+1, len(axes)):
        fig.delaxes(axes[j])
    
    plt.tight_layout()
    plt.show()

np.random.seed(42)
data = {
    'Age': np.concatenate([np.random.normal(35, 5, 500), np.random.normal(45, 8, 300)]),
    'Monthly Charges': np.concatenate([np.random.normal(65, 10, 500), np.random.normal(85, 15, 300)]),
    'Total Charges': np.concatenate([np.random.normal(2000, 500, 500), np.random.normal(3500, 800, 300)]),
    'Tenure': np.concatenate([np.random.normal(30, 10, 500), np.random.normal(15, 5, 300)]),
    'Churn Label': ['No']*500 + ['Yes']*300
}

df = pd.DataFrame(data)

numeric_cols = ['Age', 'Monthly Charges', 'Total Charges', 'Tenure']
plot_numeric_distributions(df, numeric_cols)
````

````python
# تشخیص مشتریانی که پتانسیل فروش افزایشی دارند

def add_upsell_flag(df):
    df_flagged = df.copy()
    
    avg_monthly_charge = df['Monthly Charges'].mean()
    
    median_cltv = df['CLTV'].median()
    
    df_flagged['Upsell_Flag'] = (
        (df['Churn Value'] == 0)   
        (df['Contract'] == 'Two year') 
        (df['Monthly Charges'] > avg_monthly_charge) 
        (df['CLTV'] > median_cltv) 
        (df['Payment Method'].str.contains('automatic')))
    
    return df_flagged
````

````python
# TODO: تعریف تابع برای نمایش classification report و confusion matrix
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report
import numpy as np

def plot_classification_report(y_true, y_pred, classes=['Not Churn', 'Churn'], figsize=(8, 6)):
    
    report = classification_report(y_true, y_pred, target_names=classes, output_dict=True)
    
    metrics = ['precision', 'recall', 'f1-score']
    data = []
    for cls in classes:
        row = []
        for metric in metrics:
            row.append(report[cls][metric])
        data.append(row)
    
    data.append([
        report['accuracy'],
        report['macro avg']['precision'],
        report['macro avg']['recall']
    ])
    
    plt.figure(figsize=figsize)
    sns.heatmap(
        data,
        annot=True,
        fmt='.2f',
        cmap='Blues',
        xticklabels=metrics,
        yticklabels=classes + ['Accuracy/Macro Avg']
    )
    plt.title('Classification Report')
    plt.xlabel('Metrics')
    plt.ylabel('Classes')
    plt.show()
````

````python
from sklearn.metrics import confusion_matrix

def plot_confusion_matrix(y_true, y_pred, classes=['Not Churn', 'Churn'], figsize=(8, 6)):
    cm = confusion_matrix(y_true, y_pred)
    
    plt.figure(figsize=figsize)
    sns.heatmap(
        cm,
        annot=True,
        fmt='d',
        cmap='Blues',
        xticklabels=classes,
        yticklabels=classes
    )
    plt.title('Confusion Matrix')
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.show()
````
