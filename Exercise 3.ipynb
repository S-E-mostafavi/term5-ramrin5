````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# ุชุงุจุน ุจุฑุง ุฑุณู Countplot ูุง ุฏุณุชูโุจูุฏ ุดุฏู
def plot_categorical_counts(df, categorical_columns, target_col='Churn Label'):
    """
    ุฑุณู ูููุฏุงุฑูุง Count ุจุฑุง ูุชุบุฑูุง ุฏุณุชูโุง ุจุง ุชูฺฉฺฉ ุจุฑ ุงุณุงุณ ูุชุบุฑ ูุฏู
    
    ูพุงุฑุงูุชุฑูุง:
    df (DataFrame): ุฏุชุงูุฑู ุดุงูู ุฏุงุฏูโูุง
    categorical_columns (list): ูุณุช ุณุชููโูุง ุฏุณุชูโุง
    target_col (str): ูุงู ุณุชูู ูุฏู ุจุฑุง ุชูฺฉฺฉ (ูพุดโูุฑุถ 'Churn Label')
    """
    # ุชูุธู ุณุจฺฉ ูููุฏุงุฑูุง
    sns.set(style="whitegrid")
    
    n_cols = 2
    n_rows = (len(categorical_columns) + 1) // n_cols
    
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(16, 5*n_rows))
    axes = axes.flatten()
    
    for i, col in enumerate(categorical_columns):
        if col in df.columns:
            sns.countplot(data=df, x=col, hue=target_col, ax=axes[i], 
                         palette={'Yes': 'salmon', 'No': 'lightseagreen'})
            
            axes[i].set_title(f'ุชูุฒุน {col} ุจุฑ ุงุณุงุณ {target_col}', fontsize=14)
            axes[i].set_xlabel(col, fontsize=12)
            axes[i].set_ylabel('ุชุนุฏุงุฏ', fontsize=12)
            
            axes[i].tick_params(axis='x', rotation=45)
            
            for p in axes[i].patches:
                axes[i].annotate(f'{int(p.get_height())}', 
                               (p.get_x() + p.get_width() / 2., p.get_height()),
                               ha='center', va='center', 
                               xytext=(0, 5), textcoords='offset points')
        else:
            print(f"ุณุชูู '{col}' ุฏุฑ ุฏุชุงูุฑู ูุฌูุฏ ูุฏุงุฑุฏ.")
    
    for j in range(i+1, len(axes)):
        fig.delaxes(axes[j])
    
    plt.tight_layout()
    plt.show()

np.random.seed(42)
data_size = 1000
data = {
    'Contract': np.random.choice(['Month-to-month', 'One year', 'Two year'], 
                               size=data_size, p=[0.55, 0.3, 0.15]),
    'Internet Service': np.random.choice(['DSL', 'Fiber optic', 'No'], 
                                       size=data_size, p=[0.4, 0.4, 0.2]),
    'Payment Method': np.random.choice(['Electronic check', 'Mailed check', 
                                      'Bank transfer', 'Credit card'], 
                                     size=data_size, p=[0.4, 0.2, 0.2, 0.2]),
    'Tech Support': np.random.choice(['Yes', 'No', 'No internet service'], 
                                   size=data_size, p=[0.3, 0.5, 0.2]),
    'Churn Label': np.random.choice(['Yes', 'No'], size=data_size, p=[0.3, 0.7])
}

df.loc[df['Contract'] == 'Month-to-month', 'Churn Label'] = np.random.choice(
    ['Yes', 'No'], size=sum(df['Contract'] == 'Month-to-month'), p=[0.4, 0.6])
df.loc[df['Contract'] == 'One year', 'Churn Label'] = np.random.choice(
    ['Yes', 'No'], size=sum(df['Contract'] == 'One year'), p=[0.2, 0.8])
df.loc[df['Contract'] == 'Two year', 'Churn Label'] = np.random.choice(
    ['Yes', 'No'], size=sum(df['Contract'] == 'Two year'), p=[0.05, 0.95])

df.loc[df['Internet Service'] == 'Fiber optic', 'Churn Label'] = np.random.choice(
    ['Yes', 'No'], size=sum(df['Internet Service'] == 'Fiber optic'), p=[0.4, 0.6])
df.loc[df['Internet Service'] == 'DSL', 'Churn Label'] = np.random.choice(
    ['Yes', 'No'], size=sum(df['Internet Service'] == 'DSL'), p=[0.2, 0.8])

categorical_cols = ['Contract', 'Internet Service', 'Payment Method', 'Tech Support']
plot_categorical_counts(df, categorical_cols)
````

````python
# TODO: ุชุนุฑู ุชุงุจุน ุจุฑุง ููุงุด classification report ู confusion matrix
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report
import numpy as np

def plot_classification_report(y_true, y_pred, classes=['Not Churn', 'Churn'], figsize=(8, 6)):
    
    report = classification_report(y_true, y_pred, target_names=classes, output_dict=True)
    
    metrics = ['precision', 'recall', 'f1-score']
    data = []
    for cls in classes:
        row = []
        for metric in metrics:
            row.append(report[cls][metric])
        data.append(row)
    
    data.append([
        report['accuracy'],
        report['macro avg']['precision'],
        report['macro avg']['recall']
    ])
    
    plt.figure(figsize=figsize)
    sns.heatmap(
        data,
        annot=True,
        fmt='.2f',
        cmap='Blues',
        xticklabels=metrics,
        yticklabels=classes + ['Accuracy/Macro Avg']
    )
    plt.title('Classification Report')
    plt.xlabel('Metrics')
    plt.ylabel('Classes')
    plt.show()
````

````python
from sklearn.metrics import confusion_matrix

def plot_confusion_matrix(y_true, y_pred, classes=['Not Churn', 'Churn'], figsize=(8, 6)):
    cm = confusion_matrix(y_true, y_pred)
    
    plt.figure(figsize=figsize)
    sns.heatmap(
        cm,
        annot=True,
        fmt='d',
        cmap='Blues',
        xticklabels=classes,
        yticklabels=classes
    )
    plt.title('Confusion Matrix')
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.show()
````
